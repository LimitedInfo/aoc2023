```python
game = guide[18]
```

```python
pulls = [x.split(', ') for x in game[game.find(':')+2:].split('; ')]
```

```python
limits = {'red': 12, 'green': 13, 'blue': 14}
```

```python
def possible_check(pulls, id):
    for pull in pulls:
        for num_color in pull:
            num, color = num_color.split(' ')
            if int(num) > limits[color]:
                print(pull, 'impossible')
                return False, id

    return True, id
```

```python
possible_check(pulls, int(game[5:game.find(':')]))
```

```python
possibles = []
for game in guide:
    possible, id = possible_check([x.split(', ') for x in game[game.find(':')+2:].split('; ')], int(game[5:game.find(':')]))
    if possible:
        possibles.append(id)
```

```python
sum(possibles)
```

```python
def min_check(pulls):
    max_values = {'red': 0, 'blue': 0, 'green': 0}
    for pull in pulls:

        for num_color in pull:
            num, color = num_color.split(' ')
            print(num_color)
            if max_values[color] < int(num):
                max_values[color] = int(num)
                print(max_values)

    return max_values.values()
```

```python
game = guide[18]
```

```python
pulls = [x.split(', ') for x in game[game.find(':')+2:].split('; ')]
```

```python
min_check(pulls)
```

```python
possibles = []
for game in guide:
    min_values = min_check([x.split(', ') for x in game[game.find(':')+2:].split('; ')])

    possibles.append(min_values)
```

```python
sum([math.prod(x) for x in possibles])
```
