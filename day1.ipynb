```python
sentence = '1abc2'

def get_calibration_sentence(input):
    output = []
    for sentence in input:
        result = []
        for letter in sentence:
            if 47 < ord(letter) < 58:
             result.append(letter)

        output.append(result[0] + result[-1])

    return output
```

```python
numbers_spelled = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
```

```python
def get_last_number(input):
    for sentence in input:
        result = []
        for x, letter in enumerate(reversed(sentence)):
            if 47 < ord(letter) < 58:
                result.append(letter)
                return result
            for y, number_spelt in enumerate(numbers_spelled):
                print(number_spelt, sentence[-(x+1):])
                if number_spelt in sentence[-(x+1):]:
                    result.append(y+1)
                    print('here')
                    return result
```

```python
def get_first_number(input):
    for sentence in input:
        result = []
        for x, letter in enumerate(sentence):
            if 47 < ord(letter) < 58:
                result.append(letter)
                return result
            for y, number_spelt in enumerate(numbers_spelled):
                print(number_spelt, sentence[:x+1])
                if number_spelt in sentence[:x+1]:
                    result.append(y+1)
                    print('here')
                    return result
```

```python
def get_first_get_last_combine(input):
    first = get_first_number(input)
    last = get_last_number(input)
    return str(first[0]) + str(last[0])
```

```python
out = get_first_get_last_combine(['eightwothree'])
```

```python
results = []
for input in guide:
    res = get_first_get_last_combine([input])
    results.append(res)
```
